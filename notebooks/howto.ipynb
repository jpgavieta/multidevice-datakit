## Loading All Data

The function `load_data()` automatically triggers the etl (extract, transform, load) pipeline to prepare all available device data for analysis. 
This is the one-liner command to load everything. 

**Note**: This pipeline is currently extracts from a mounted Proton Drive folder, it can be upgraded to API calls per `device_id` instead.

```python
import sys
sys.path.insert(0, "..") # run at repo root

from src.etl.load import load_data

MOUNT_PATH = "/home/yul/mnt/proton-data"

data = load_data(MOUNT_PATH)
```

## Navigating Devices, Dataframes, and Variables

The function `display_loaded_data()` quickly shows the available dataframes (`df`) per device (`device_id`), and its shape (columns nos. and rows nos.) and range (start datetime to end datetime).

```python
from src.etl.load import display_loaded_data

display_loaded_data(data) # Displays all of the loaded available data, devices, and device types

display_loaded_data({"Atmotube": data["Atmotube"]}) # Displays loaded data from a specific device type
```

Right now, the `data` is holding all data for multiple device types (`device_type`) and device IDs (`device_id`).

Navigate to a specific device (`device_id`), its dataframes (`df`), and variables (`cols`).

```python
# See which device types are available 
list_device_types = data.keys() 
print(list_device_types)

# See available device IDs, per type
list_device_ids = data["Atmotube"].keys()
print(list_device_ids)

# Get a specific dataframe
device1_pm_df = data["Atmotube"]["C3CBE16AE294_01-May-2026_12-Jun-2026"]["data"]["pm"]["df"]
print(device1_pm_df)

# See available variables in that dataframe
device1_pm_cols = data["Atmotube"]["C3CBE16AE294_01-May-2026_12-Jun-2026"]["data"]["pm"]["cols"]
print(device1_pm_cols)

# Get a specific variable 
device1_pm2_5 = data["Atmotube"]["C3CBE16AE294_01-May-2026_12-Jun-2026"]["data"]["pm"]["df"]["pm2_5_ugm3_atm"]
print(device1_pm2_5)
```
